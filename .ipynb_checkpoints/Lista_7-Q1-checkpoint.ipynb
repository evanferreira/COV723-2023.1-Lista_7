# Espectro de Pierson-Moskowitz Modificado

```python
import numpy as np
from matplotlib import pyplot as plt
import random
from scipy.signal import find_peaks, argrelextrema
from scipy import integrate
import pandas as pd

Hs = 7.8 
Tz = 9.2
T1 = 1.086*Tz
dw = 0.01
w = np.arange(0.01, np.pi/2, dw)
S_w = ((173*Hs**2)/(T1**4))*(w**(-5))*np.exp(-(692/(T1**4))*w**(-4))
N = 3

plt.figure(figsize=(12,6))
plt.plot(w, S_w, label=r'$S_{\zeta}(\omega)$')
plt.grid()
plt.legend()
plt.title(r'Espectro de Pierson-Moskowitz Modificado - $H_s$ = {Hs} e $T_p$ = {T1:2f}'.format(Hs = Hs, T1=T1))
plt.xlabel(r'$\omega$, rad/s')
plt.ylabel(r'$S_{\zeta}(\omega)$, m/s', loc='top', labelpad=-10, rotation=0)
plt.show()

dw = w[1]-w[0]
t = np.arange(0, 200, 0.01)
zeta_w = np.sqrt(2*S_w*dw)
zeta_t = np.zeros_like(t)
T_zeta = np.zeros(N)
sig_zeta = np.zeros(N)

for n in range(0, N):
    for elev, freq in zip(zeta_w, w):
        fase = random.uniform(0, 2*np.pi)
        z = elev*np.cos(-freq*t + fase)
        zeta_t += z
    plt.figure(figsize=(12,6))
    peaks, _ = find_peaks(zeta_t, None)
    valleys = argrelextrema(zeta_t, np.less)
    plt.plot(t, zeta_t)
    plt.plot(t[peaks], zeta_t[peaks] , "x", color='g')
    plt.plot(t[valleys], zeta_t[valleys] , "x", color='r')
    plt.axhline(y=0, color='gray')
    plt.title(f'Srie temporal - Elevao da onda - {n+1} simulao')
    plt.xlabel('t, s')
    plt.ylabel(r'$\zeta(t)$, m', loc='top', labelpad=-10, rotation=0)
    plt.grid()
    plt.show()

    
# np.mean(zeta_t)
# np.std(zeta_t)

# peaks, _ = find_peaks(zeta_t, None)
# valleys = argrelextrema(zeta_t, np.less)

# fig, ax = plt.subplots(figsize=(12,6))
# plt.plot(t, zeta_t)
# plt.plot(t[peaks], zeta_t[peaks] , "x", color='g')
# plt.plot(t[valleys], zeta_t[valleys] , "x", color='r')
# plt.axhline(y=0, color='gray')
# plt.grid()
# plt.show()
# print(np.mean(zeta_t[peaks]), np.std(zeta_t[peaks]), np.mean(zeta_t[valleys]), np.std(zeta_t[valleys]))
# print(integrate.simpson(S_w, w), integrate.simpson(S_w*w, w))
# print(np.sqrt(integrate.simpson(S_w, w)))
# 2*np.pi*np.sqrt((integrate.simpson(S_w, w)/integrate.simpson(S_w*w**2, w)))
```
